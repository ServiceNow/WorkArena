```python
import ast
import json
import pandas as pd

workarena_dash_task_file = "./src/browsergym/workarena/data_files/task_configs/dashboard_retrieval_value_task.json"
```

```python
# Load failure analysis results
data = pd.read_csv("task_failure_ranking.csv")
data.config = data.config.map(lambda x: ast.literal_eval(x)["config"])
data = data.loc[data.status == "confirmed_broken"]
print(data.shape)
data.head()
```

```python
# Load all workarena tasks
all_dash_tasks = json.load(open(workarena_dash_task_file, "r"))
len(all_dash_tasks)
```

```python
all_dash_tasks[0]
```

```python
# Remove any task that was considered broken
revised_tasks = []
for task in all_dash_tasks:
    if data.config.map(lambda x: x == task).any():
        print("yes")
    else:
        revised_tasks.append(task)
json.dump(revised_tasks, open(workarena_dash_task_file, "w"))
```
